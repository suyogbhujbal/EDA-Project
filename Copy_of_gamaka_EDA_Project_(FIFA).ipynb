```python
fdf = fdf.rename(columns={'Value':'Value (in €)','Wage':'Wage(in €)',
                                  'Release Clause':'Release Clause(in €)'})
fdf.columns
```

```python
fdf['Value (in €)'].unique()
```

```python
fdf['Value (in €)'] = fdf['Value (in €)'].astype(str) 
fdf['Value (in €)'] = fdf['Value (in €)'].apply(lambda x: x.replace('€','')) 
fdf['Value (in €)'] = fdf['Value (in €)'].apply(lambda y: y.replace('.',''))
fdf['Value (in €)'] = fdf['Value (in €)'].apply(lambda z: z.replace('K','000'))
fdf['Value (in €)'] = fdf['Value (in €)'].apply(lambda w: w.replace('M','000000'))
fdf['Value (in €)'] = fdf['Value (in €)'].astype(int)
```

```python
fdf['Wage(in €)'].unique()
```

```python
fdf['Wage(in €)'] = fdf['Wage(in €)'].astype(str) 
fdf['Wage(in €)'] = fdf['Wage(in €)'].apply(lambda x: x.replace('€','')) 
fdf['Wage(in €)'] = fdf['Wage(in €)'].apply(lambda z: z.replace('K','000'))
fdf['Wage(in €)'] = fdf['Wage(in €)'].astype(int)
```
